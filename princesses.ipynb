```python
with open('datasets_12dancingprincesses.txt', 'r') as file:
    data = file.read()
    
    
```

```python
# change to lower case
data=data.lower()
```

```python
data # display
```

```python
word_tokenize(data) # tokenize data so it is split
```

```python
sent_tokenize(data) #sentence tokenize by comma split
```

```python
result_tknz = word_tokenize(data)  #assign to a variable 
```

```python
result_tknz[:-20]  #show a few
```

```python
#number of tokens in list before puntuation removal
len(result_tknz)
```

```python
#remove the puntuation tokens from the list
for word in result_tknz:
    if word in punctuation:
        result_tknz.remove(word)
```

```python
#number of tokens in list after puntuation removal
len(result_tknz)
```

```python
punctuation
```

```python
#remove filler words
eng_stopwords = stopwords.words('english')
new_words = []  #list to hold new words

for word in result_tknz:
    if word not in eng_stopwords:
        new_words.append(word)
```

```python
new_words #list without punctuation and stop words
```

```python
for word in new_words:
    if word in ['’', '‘']:
        new_words.remove(word)
```

```python
fd_nw = FreqDist(new_words)
fd_nw.most_common(10)
```
